```python
info_columnas = pd.read_csv("../../data/info_columnas/info_columnas.csv")
```

```python
apariciones = info_columnas.groupby("columna").agg({"dataframe" : "count"}).sort_values(by="dataframe", ascending=False).reset_index()
```

```python
apariciones.columns = ["columna", "apariciones"]
```

```python
info_columnas = pd.merge(info_columnas, apariciones, how="inner", on="columna")
```

```python
info_columnas.sort_values(by=["apariciones", "columna"], ascending=[False, True], inplace=True)
```
