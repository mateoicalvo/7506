# Installs - Carga de datos

```python
installs = pd.read_csv("../../data/tp2/installs.csv")
```

```python
installs.memory_usage(deep=True).sum() / 1024**2
```

```python
exploracion = installs.dtypes.to_frame()
```

```python
exploracion.columns = ["dtype"]
```

```python
exploracion["tiene_nulos"] = installs.isna().any()
```

```python
exploracion["todos_nulos"] = installs.isna().all()
```

```python
exploracion["cantidad_nulos"] = len(installs) - installs.count()
```

```python
exploracion["porcentaje_nulos"] = exploracion["cantidad_nulos"] / len(installs)
```

```python
exploracion["porcentaje_nulos"] = exploracion["porcentaje_nulos"].map(lambda x: round(x*100, 2))
```

```python
exploracion["maximo"] = installs.max()
```

```python
exploracion["minimo"] = installs.min()
```

```python
exploracion["valores_unicos"] = installs.nunique()
```

```python
exploracion["porcentaje_valores_unicos"] = (exploracion["valores_unicos"] / len(installs)) *100 
exploracion["porcentaje_valores_unicos"] = exploracion["porcentaje_valores_unicos"].map(lambda x: round(x, 2))
```
